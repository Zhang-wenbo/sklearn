### 1.归一化

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = [[2, 1], [3, 1], [1, 4], [2, 6]]
```

```python
# 定义归一化类的对象
scaler = MinMaxScaler(feature_range=(-1,1))
# 将缩放器应用于特征上
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

### 2.标准化

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = [[2, 1], [3, 1], [1, 4], [2, 6]]
```

```python
# 定义标准化类的对象
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```
