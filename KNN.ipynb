### 1.分类算法

```python
from sklearn.neighbors import KNeighborsClassifier # 引入KNN分类
import numpy as np
import matplotlib.pyplot as plt

from B站课程演示.动力学模拟绘图.MD绘图 import colors
```

```python
# 准备数据
X = np.array([[2, 1], [3, 1], [1, 4], [2, 6]])  # 特征4*2的矩阵
y = np.array([0, 0, 1, 1])  # 分类标签
print(X)
```

```python
# 定义KNN分类模型
knn = KNeighborsClassifier(n_neighbors=2)  # KNN分类模型，K值为2
knn.fit(X, y)  # 模型训练
X_class = knn.predict([[4, 9]])  # 预测
print(X_class)
```

```python
# 可视化
fig, ax = plt.subplots()
ax.axis('equal')
# 使用布尔索引将两类点分开
X_1 = X[y == 0]
X_2 = X[y == 1]
colors = ["C0", "C1"]
plt.scatter(X_1[:, 0], X_1[:, 1], c=colors[0], marker='o')
plt.scatter(X_2[:, 0], X_2[:, 1], c=colors[1], marker='x')
# 新点
x_color = colors[0] if X_class == 0 else colors[1]
plt.scatter(x=[4], y=[9], c=x_color, marker='*', s=100)
plt.show()
```

### 2.回归算法

```python
from sklearn.neighbors import KNeighborsRegressor # KNN回归
```

```python
# 准备数据
X = [[2, 1], [3, 1], [1, 4], [2, 6]]  # 特征4*2的矩阵
y = [0, 0, 1, 1]  # 分类标签
```

```python
# KNN回归模型
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(X, y)
# 预测
x = [[4,9]]
y_predict = knn.predict(x)
print(y_predict)
```
