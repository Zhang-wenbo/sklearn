## 1.函数绘图

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义函数y=sin(x)
x = np.arange(0, 6, 0.1) # 从0到6，步长为0.1
y = np.sin(x)
```

```python
# 绘图
plt.plot(x, y)
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 函数与导函数 y'=cos(x)
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend() # 显示图例
plt.show()
```

## 2.向量

```python
import numpy as np
# 定义向量
x = np.array([2, 5, 8])
print(x)
print(x.shape)
```

```python
# 向量转置
print(x.T)
print(x.T.shape)
```

```python
# 向量相加
y = np.array([1,3,7])
print(x+y)
```

```python
# 向量与标量相乘
print(x*3)
```

```python
# 向量与向量相乘(直接用乘法，得到的是对应位置相乘)
print(x*y)
# 点乘([2,5,8]·[1,3,7]=2*1+5*3+8*7=73)
print(x.dot(y))
print(x @ y)
```

```python
# 计算范数
l1_norm = np.linalg.norm(x, ord=0)
print(l1_norm)
l2_norm = np.linalg.norm(x, ord=1)
print(l2_norm)
```

## 3.矩阵

```python
import numpy as np
A = np.array([[1,2,3], [4,5,6]])
B = np.array([[1,3,7], [5,0,2]])
print(A)
print(A.shape)
```

```python
# 转置
print(A.T)
```

```python
# 矩阵相加
print(A+B)
```

```python
# 矩阵与标量相乘
print(A*3)
```

```python
# 矩阵与向量相乘
x = np.array([2,3,4])
print(A)
print(A * x)
```

```python
# 矩阵与矩阵相乘
print(A)
print(B)
print(A * B) # 这是对应位置相乘,要求两个矩阵形状相同
print(A.dot(B.T))
print(A @ B.T) # 这是真正的矩阵乘法
```

```python
# 求逆
C = np.array([[1,2], [3,5]])
C_inv = np.linalg.inv(C)
print(C_inv)
C_inv = np.around(C_inv).astype(int)
print(C_inv)
print(C_inv @ C) # 得到单位矩阵
```

```python
# 矩阵的向量化
print(A)
print(A.flatten())
```

```python
# 矩阵内积
print('<A,B> = ',np.sum(A*B))
```

```python
# 克罗内克积
print(np.kron(A,B))
```

```python
# 张量
tensor = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(tensor)
```

```python
# 计算梯度
f = np.array([1,2,4,7,11,16])
df = np.gradient(f)
print(df)
```
