# 逻辑门电路

## 与门

```python
def AND(x1,x2):
    w1,w2,theta = 0.5,0.5,0.7
    res = x1*w1+x2*w2
    if res > theta:
        return 1
    else:
        return 0
```

```python
# 测试
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

## 与门-新形式

```python
import numpy as np
def AND_np(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.7
    # 直接用矩阵运算的形式计算结果
    res = w @ x + b
    if res > 0:
        return 1
    else:
        return 0
```

```python
# 测试
print(AND_np(0,0))
print(AND_np(0,1))
print(AND_np(1,0))
print(AND_np(1,1))
```

## 与非门

```python
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    # 直接用矩阵运算的形式计算结果
    res = w @ x + b
    if res > 0:
        return 1
    else:
        return 0
```

```python
# 测试
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

## 或门

```python
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.2
    # 直接用矩阵运算的形式计算结果
    res = w @ x + b
    if res > 0:
        return 1
    else:
        return 0
```

```python
# 测试
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

## 异或门-利用多层感知机实现

```python
import numpy as np

def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w * x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(0, 0))  # 0
print(XOR(1, 0))  # 1
print(XOR(0, 1))  # 1
print(XOR(1, 1))  # 0
```
